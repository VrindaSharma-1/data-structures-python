### Infix to PostFix

```python
import re
class Stack:
    def __init__(self):
        self.items=[]
        
    def push(self,item):
        self.items.insert(0,item)
                

    def pop(self):
        return self.items.pop(0)
                    
                
    def isEmpty(self):
        return (self.items==[])
    def peek(self):
        return self.items[0]
    def size():
        return len(self.items)
    

    
def iTop(infixexpr):
    print("Input :",infixexpr)
    prec={}
    prec["^"]=4
    prec["*"]=3
    prec["/"]=3
    prec["+"]=2
    prec["-"]=2
    prec["("]=1
    stackObject=Stack()
    postfixArray=[]
    tokenList=infixexpr.split()
    print(tokenList)
    
    for token in tokenList[0]:
        if re.search(r"^\w+",token):
#         if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfixArray.append(token)
#             print(postfixArray)
            
        elif token =='(':
            stackObject.push(token)
        elif token==')':
            topToken=stackObject.pop()
            while topToken!='(':
                postfixArray.append(topToken)
                topToken=stackObject.pop()
        else:
            while (not stackObject.isEmpty()) and (prec[stackObject.peek()]>=prec[token]):
                postfixArray.append(stackObject.pop())
#                 print(prec[token])
            stackObject.push(token)
            
    while not stackObject.isEmpty():
        postfixArray.append(stackObject.pop())
#     print(postfixArray)
    return " ".join(postfixArray)

string ='(A-B+(C*D))'
iTop(string)
        
    
```

### Infix to Prefix


```python
from __future__ import division
import random


OPERATORS = set(['+', '-', '*', '/', '(', ')'])
PRIORITY = {'+':1, '-':1, '*':2, '/':2}

def infix_to_prefix(formula):
    op_stack = []
    exp_stack = []
    for ch in formula:
        if not ch in OPERATORS:
            exp_stack.append(ch)
        elif ch == '(':
            op_stack.append(ch)
        elif ch == ')':
            while op_stack[-1] != '(':
                op = op_stack.pop()
                a = exp_stack.pop()
                b = exp_stack.pop()
                exp_stack.append( op+b+a )
            op_stack.pop() # pop '('
        else:
            while op_stack and op_stack[-1] != '(' and PRIORITY[ch] <= PRIORITY[op_stack[-1]]:
                op = op_stack.pop()
                a = exp_stack.pop()
                b = exp_stack.pop()
                exp_stack.append( op+b+a )
            op_stack.append(ch)
    
    # leftover
    while op_stack:
        op = op_stack.pop()
        a = exp_stack.pop()
        b = exp_stack.pop()
        exp_stack.append( op+b+a )
#     print(exp_stack[-1])
    return exp_stack

infix_to_prefix('((A+B)*C-D)')
 
```

### LinkedList

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

        
class LinkedList:
    def __init__(self):
        self.root=None
        
    def push(self, data):
        newnode=Node(data)
        newnode.next=self.root
        self.root=newnode
        print("Added ",data,"to the linked list!")
    def pop(self):
        temp=self.root
        self.root=self.root.next
        print("Popped :",temp.data)
    def isEmpty(self):
        return True if self.root.next is None else False
    
    def insertAfter(self,node,data):
        insertNode=Node(data)
        insertNode.next=node.next
        node.next=insertNode
        print("Inserted ",data,"after",node.data,"in the linked list")
    def printlist(self):
        temp=self.root
        while temp:
            print(temp.data, end=" ")
            temp=temp.next
    
    def appendData(self,data):
        newnode=Node(data)
        last=self.root
        while last.next!=None:
            last=last.next
        last.next=newnode
        newnode.next=None
        
    def delete(self,index):
        prev=self.root
        
        while prev.next!=index:
            if(index==self.root):
                self.root=self.root.next
                return
            else:
                prev=prev.next
        prev.next=index.next
        index.next=None
        
    def getLength(self):
        length=0
        node=self.root
        
        while node!=None:
            length+=1
            node=node.next
            
        return length
    
    def getNode(self,index):
#         get nth node from the end of the linked list
        length=LinkedList.getLength(self)   
        getNode=self.root
        if index>length:
            print("Location is greater than the length of the linked list")
            return
        for i in range(length-index):
            getNode=getNode.next
        return getNode.data
            
    def swap(self,k): 
#         swap kth node from the beginning with the kth node from the end
        length=LinkedList.getLength(self) 
        beg=self.root
        end=self.root
        for i in range(k-1):
            beg=beg.next
            
        for i in range(length-k):
            end=end.next
        
        temp=beg.data
        beg.data=end.data
        end.data=temp
        
    def removeEveryKth(self,k):
        
        length=LinkedList.getLength(self)
        node=self.root

        for i in range(int(length/k)):
            for i in range(k-1):
                prev=node
                node=node.next
            prev.next=node.next
            node=prev.next
    def compare(self):
        #sum of nodes inalinked list which are greater than next node
        length=LinkedList.getLength(self)
        node=self.root
        sum=0
        
        for i in range(length-3):
            a=node.data
            b=node.next.data
            c=node.next.next.data
            if(a+b>c):
                sum=sum+a+b
            
            node=node.next
        return sum
    def midInsert(self,data):
        length=LinkedList.getLength(self)
        node=self.root
        newnode=Node(data)
        for i in range(int(length/2)-1):
            node=node.next

        newnode.next=node.next   
        node.next=newnode
        
    def balance(self):
        length=LinkedList.getLength(self)
        begSum=0
        endSum=0
        totalSum=0
        beg=self.root
        node=self.root
                          
        while node!=None:
            totalSum=totalSum+node.data
            node=node.next
           
            
                
        print("\nTotal",totalSum)
        
        for i in range(length):
            begSum = begSum + beg.data
            endSum = totalSum - begSum
            
            if (begSum==endSum):
                print(endSum)
                print(begSum)
                return beg.data
            elif beg.next==None:
                return -1
            else:
                beg=beg.next
                
#create an object of LinkedList class    
llist=LinkedList()

#push some elements to the linked list
llist.push(1)
llist.push(2)
llist.push(3)

#pop the last pushed element
llist.pop()

# insert an element after given node index
llist.insertAfter(llist.root.next,4)

#print the linkedlist
llist.printlist()

#Append an element at the end of the linkedlist
llist.appendData(5)
print()
llist.printlist()

# delete an element from the linked list for a given index
llist.delete(llist.root.next)
print()
llist.printlist()
llist.appendData(8)
print()
llist.printlist()
print()

# Get nth node value from the end of the linkedList
print(llist.getNode(1))

llist.printlist()

#Swap the kth element from the beginning with the kth elemen from the end
llist.swap(2)
print()
llist.printlist()

# remove every kth element from the linkedlist
llist.removeEveryKth(2)
print()
llist.printlist()
print()
llist.push(1)
llist.push(2)
llist.push(3)
print()
llist.appendData(1)
llist.appendData(2)
llist.appendData(3)
print()

# Return the sum of all the nodes whose sum is greater than the next node
print(llist.compare())
print()
llist.printlist()

#Insert a node in the middle of the linkedlist
llist.midInsert(7)
print()
llist.printlist()

llist.midInsert(7)
print()
llist.insertAfter(llist.root.next,2)
llist.printlist()

# return the node before which the sum of all nodes is equal to that of the nodes after it
llist.balance()
```
